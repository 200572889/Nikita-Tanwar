import math

def triangleArea(a, b, c):
    # Calculate the semi-perimeter (s)
    s = (a + b + c) / 2

    # Calculate the area using Heron's formula
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    return area

# Example usage:
result = triangleArea(2, 2, 2)
print(result)

def inside(x, y, x1, y1, x2, y2):
    # Check if (x, y) lies inside the rectangle defined by (x1, y1) and (x2, y2)
    return x1 <= x <= x2 and y1 <= y <= y2

def pig(word):
    # Convert the word to lowercase to ensure case insensitivity
    word = word.lower()

    # Check if the word starts with a vowel
    if word[0] in 'aeiou':
        return word + 'way'
    else:
        # Find the index of the first vowel
        for i in range(len(word)):
            if word[i] in 'aeiou':
                # Move the consonants before the first vowel to the end and add 'ay'
                return word[i:] + word[:i] + 'ay'

def bldcount(filename):
    # Create a dictionary to store the count of each blood type
    blood_type_count = {'A': 0, 'B': 0, 'AB': 0, 'O': 0, 'OO': 0}

    try:
        # Open and read the file
        with open(filename, 'r') as file:
            for line in file:
                # Split the line into words and process each word
                words = line.split()
                for word in words:
                    # Check if the word is a valid blood type
                    if word in blood_type_count:
                        # Increment the count for the corresponding blood type
                        blood_type_count[word] += 1

        # Print the results
        for blood_type, count in blood_type_count.items():
            if count > 0:
                if count == 1:
                    print(f"There is one patient of blood type {blood_type}.")
                else:
                    print(f"There are {count} patients of blood type {blood_type}.")
            else:
                print(f"There are no patients of blood type {blood_type}.")

    except FileNotFoundError:
        print(f"File '{filename}' not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Call the function with the filename 'bloodtype.txt'
bldcount('bloodtype.txt')

def curconv(currency, amount):
    exchange_rates = {}
    
    try:
        # Read currency exchange rates from 'currencies.txt'
        with open('currencies.txt', 'r') as file:
            lines = file.readlines()
            
            # Extract exchange rates and store them in a dictionary
            for i in range(0, len(lines), 3):
                currency_code = lines[i].strip()
                rate = float(lines[i + 1])
                exchange_rates[currency_code] = rate

        # Check if the input currency code is valid
        if currency in exchange_rates:
            # Convert the amount to US dollars
            usd_amount = amount / exchange_rates[currency]
            return usd_amount
        else:
            return "Invalid currency code."

    except FileNotFoundError:
        return "File 'currencies.txt' not found."
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Example usage:
usd_amount_eur = curconv('EUR', 100)
usd_amount_jpy = curconv('JPY', 100)

print(usd_amount_eur)
print(usd_amount_jpy)
